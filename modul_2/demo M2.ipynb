## Pandas Package

```python
# Membaca dataset demo
df_demo = pd.read_csv('demo_dataset_m2.csv')

# Melihat 5 baris pertama
print("Tampilan Awal Data:")
display(df_demo.head())

# Mengecek daftar kolom dan tipe data
print("\nInformasi Struktur Data:")
df_demo.info()
```

```python
# 1. Konversi Kolom Tanggal menjadi tipe Datetime
df_demo['tanggal'] = pd.to_datetime(df_demo['tanggal'])

# 2. Menangani Missing Values (NaN)
# Mengisi nilai kosong pada polutan dengan angka 0 (atau bisa menggunakan rata-rata/mean)
df_demo['pm10'] = df_demo['pm10'].fillna(0)

# 3. Menghapus kolom yang terlalu banyak kosong (seperti pm25 jika tidak diperlukan)
df_clean = df_demo.drop(columns=['pm25'])

print("Data setelah pembersihan awal:")
df_clean.head()
```

```python
# Mencari data di mana kategori udara 'TIDAK SEHAT'
kondisi_buruk = df_clean[df_clean['categori'] == 'TIDAK SEHAT']

print(f"Ditemukan {len(kondisi_buruk)} hari dengan kualitas udara TIDAK SEHAT.")
kondisi_buruk[['tanggal', 'stasiun', 'max', 'critical']].head()
```

```python
# Menghitung rata-rata nilai polutan maksimal per stasiun
analisis_stasiun = df_clean.groupby('stasiun')['max'].mean().sort_values(ascending=False)

print("Rata-rata Nilai Maksimal Polutan per Stasiun:")
print(analisis_stasiun)
```
