```python
import pandas as pd
import glob
import os

# 1. Tentukan lokasi folder tempat file Excel berada
# Ganti path ini sesuai lokasi kamu, misal: 'data_transaksi/*.xlsx'
folder_path = 'Dataset_M2/*.xlsx'
files = glob.glob(folder_path)

print(f"Ditemukan {len(files)} file.")

# 2. Loop untuk mengintip nama kolom di setiap file
print("\n--- MENGINTIP STRUKTUR KOLOM ---")
unique_columns = set()

for file in files:
    # Trik: nrows=0 artinya kita cuma baca header, jadi SANGAT CEPAT
    df_temp = pd.read_excel(file, nrows=0)

    # Simpan nama kolom unik untuk referensi kita
    current_cols = set(df_temp.columns)
    unique_columns.update(current_cols)

    # Opsional: Print nama file jika kolomnya mencurigakan/berbeda
    # (Hanya untuk debugging mata manusia)
    print(f"File: {os.path.basename(file)}")
    print(f"   Kolom: {list(df_temp.columns)}")

print("\n--- SEMUA NAMA KOLOM UNIK YANG DITEMUKAN ---")
print(unique_columns)
```

```python
# --- KONFIGURASI ---

# 1. Tentukan kolom yang ingin kita pertahankan (Standard Columns)
target_columns = [
    'order_id',
    'Waktu Pesanan Dibuat',
    'Status Pesanan',
    'Total Berat',
    'Alasan Pembatalan',
    'product_category',
    'Total Pembayaran'
]

# 2. Kamus Mapping untuk menyamakan nama kolom (Standardization)
# Sesuaikan 'nama_lama' jika Anda menemukan variasi lain di file tertentu
column_mapping = {
    'Order ID': 'order_id',
    'Waktu Pesanan': 'Waktu Pesanan Dibuat',
    'Tanggal Dibuat': 'Waktu Pesanan Dibuat',
    'Creation Time': 'Waktu Pesanan Dibuat',
    # Tambahkan mapping lain jika ada nama kolom yang berbeda untuk tujuan yang sama
}

# --- PROSES PENGGABUNGAN (Tahap 1 Poin 3) ---
all_df = []

for file in files:
    # Membaca file
    df_temp = pd.read_excel(file)

    # A. Standardisasi Nama Kolom (Inkonsistensi Nama)
    df_temp = df_temp.rename(columns=column_mapping)

    # B. Filter Kolom (Inkonsistensi Jumlah Kolom)
    # Kita hanya mengambil kolom yang ada di target_columns
    # Jika ada kolom yang tidak ada di file tersebut, kita abaikan atau beri nilai kosong
    existing_target_cols = [
        col for col in target_columns if col in df_temp.columns]
    df_clean = df_temp[existing_target_cols].copy()

    all_df.append(df_clean)

# C. Gabungkan menjadi df_master
df_master = pd.concat(all_df, ignore_index=True)

# D. Tampilkan informasi struktur untuk pembuktian (Instruksi Poin 3)
print("--- STRUKTUR DF_MASTER BERHASIL DIGABUNG ---")
# Menampilkan informasi kolom dan jumlah data non-null [cite: 25]
print(df_master.info())
print("\nContoh Data:")
display(df_master.head())
```
