```python
#tahap 1: Import Library & standarisasi data
import pandas as pd
import os

folder_path = r"G:\Project\VINIX7\modul_2\Dataset_M2"

unique_columns = set()

for file in os.listdir(folder_path):
    if file.endswith(".xlsx"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_excel(file_path, nrows=0)

        for col in df.columns:
            unique_columns.add(col)

print("Daftar semua kolom unik:\n")

for col in sorted(unique_columns):
    print(col)
```

```python
column_frequency = {}

for file in os.listdir(folder_path):

    if file.endswith(".xlsx"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_excel(file_path, nrows=0)

        for col in df.columns:
            if col in column_frequency:
                column_frequency[col] += 1
            else:
                column_frequency[col] = 1

print("\nFrekuensi kemunculan kolom:\n")

for col, count in sorted(column_frequency.items(), key=lambda x: x[1], reverse=True):
    print(f"{col} : {count}")
```

```python
standard_columns = [
    "order_id",
    "product_category",
    "Status Pesanan",
    "Alasan Pembatalan",
    "Opsi Pengiriman",
    "Metode Pembayaran",
    "Jumlah",
    "Returned quantity",
    "Total Diskon",
    "Total Berat",
    "Ongkos Kirim Dibayar oleh Pembeli",
    "Estimasi Potongan Biaya Pengiriman",
    "Total Pembayaran",
    "Perkiraan Ongkos Kirim",
    "Kota/Kabupaten",
    "Provinsi",
    "order_created_time"
]
```

```python
filtered_data = []

for file in os.listdir(folder_path):

    if file.endswith(".xlsx"):

        file_path = os.path.join(folder_path, file)

        df = pd.read_excel(file_path)

        # standardisasi nama kolom waktu
        df = df.rename(columns={
            "Waktu Pesanan Dibuat": "order_created_time",
            "Waktu Pembayaran Dilakukan": "payment_time",
            "Waktu Pesanan Selesai": "order_finished_time"
        })

        cols = [c for c in standard_columns if c in df.columns]

        df_filtered = df[cols]

        filtered_data.append(df_filtered)
```

```python
df_master = pd.concat(filtered_data, ignore_index=True)

print(df_master.info())
print(df_master.shape)

df_master.head()
```

```python
# Tahap 2: Data Cleaning
df_master["Alasan Pembatalan"] = df_master["Alasan Pembatalan"].fillna("Tidak Batal")
print(df_master["Alasan Pembatalan"].value_counts())
```

```python
# Versi lebih robust untuk menghapus "gr"
df_master["Total Berat"] = (
    df_master["Total Berat"]
    .astype(str)
    .str.replace("gr", "", case=False, regex=False)
    .str.strip()
)
df_master["Total Berat"] = pd.to_numeric(
    df_master["Total Berat"], errors="coerce")
```

```python
df_master["order_created_time"] = pd.to_datetime(df_master["order_created_time"], errors="coerce")

print(df_master["order_created_time"].dtype)
```

```python
print("Jumlah baris sebelum hapus duplikat:", df_master.shape[0])

df_master = df_master.drop_duplicates()

print("Jumlah baris setelah hapus duplikat:", df_master.shape[0])

print(df_master.info())
```

```python
# Tahap 3: Data Wrangling & Manipulasi

df_sukses = df_master[df_master["Status Pesanan"].isin(["Selesai", "Telah Dikirim"])]

print(df_sukses.shape)
df_sukses.head()
```

```python
top_product_category = (
    df_sukses
    .groupby("product_category")["Jumlah"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_product_category)
```

```python
top_product_category = (
    df_sukses
    .groupby("product_category", as_index=False)["Jumlah"]
    .sum()
    .sort_values(by="Jumlah", ascending=False)
    .head(10)
)

print(top_product_category)
```
